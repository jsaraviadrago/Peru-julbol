```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the shapefile
# Download from here: https://www.naturalearthdata.com/downloads/110m-cultural-vectors/ (whole country)
shapefile_path = 'C:\\Users\\JuanCarlosSaraviaDra\\Downloads\\ne_110m_admin_0_countries\\ne_110m_admin_0_countries.shp'
world = gpd.read_file(shapefile_path)

```

```python
# Filter for Peru
peru = world[world['NAME'] == 'Peru']

# Plot the map
peru.plot()
plt.title('Map of Peru')
plt.show()
```
